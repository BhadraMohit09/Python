### 01) WAP to find sum of all the elements in List.

```python
myList = []
size = int(input("Enter the size of list:: "))

print("Enter elements:: ")
for i in range(0, size+1):
  num = int(input())
  myList.append(num)

total = sum(myList)
print("Sum of elements is ==>",total)
```

### 03) WAP to split the List into two and append the first part to the end.

```python
myList = []
limit = int(input("Enter limit:: "))
print("Enter elements:: ")
for i in range(0, limit+1):
  num = int(input())
  myList.append(num)

mid = len(myList)//2
part1 = myList[:mid]
part2 = myList[mid:]

result = part2 + part1
print("Splitted List is ==> ",result)
```

### 07) Python program to remove multiple elements from a list using list comprehension

```python
myList = []
removeList = []
limit = int(input("Enter limit of elements: "))
print("Enter elements: ")
for i in range(0,limit):
  num = int(input())
  myList.append(num)
print("List => ",myList)
total = int(input("Enter total numbers of element to remove: "))
for i in range(0,total):
  num1 = int(input())
  removeList.append(num1)
print("Elements to remove are: ",removeList)


modifiedList =  [x for x in myList if x not in removeList]
print("List after removing elements=> ",modifiedList)
```

### 01) WAP to count Even and Odd numbers in a List.

```python
myList = []
evenCount = 0
oddCount = 0
limit = int(input("Enter limit:: "))
for i in range(0, limit):
  num = int(input())
  myList.append(num)
print("List is => ",myList)

for i in range(0,limit):
  if(myList[i]%2==0):
    evenCount += 1
  else:
    oddCount += 1
print("Total Even numbers in list => ",evenCount)
print("Total Odd numbers in list => ",oddCount)
```

### 02) Python program to find N largest and smallest elements from the list

```python
myList = []
limit = int(input("Enter limits:: "))

for i in range(0, limit):
  num = int(input())
  myList.append(num)
print("List is => ", myList)
N = int(input("Enter N elements for inputs:: "))

myList.sort()

largest = myList[-N:]
smallest = myList[:N]
print("Total",N,"largest elements => ",largest)
print("Total",N,"smallest elements => ",smallest)
```

### 03) WAP to print duplicates from a list of integers

```python
myList = []
limit = int(input("Enter limit:: "))
for i in range(0, limit):
  num = int(input())
  myList.append(num)

mySet = set()
duplicates = []

for i in myList:
  if i in mySet:
    duplicates.append(i)
  else:
    mySet.add(i)
print("Duplicates are = > ",duplicates)
```
